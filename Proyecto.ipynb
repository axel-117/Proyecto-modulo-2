**Atleta con mayores ganancias por deporte por década**

```python
earn=df1.iloc[:,[1,5,6,7]]
earn
```

```python
d1= earn[earn["Year"] <2000]

d2= earn[earn["Year"] > 1999 & (earn["Year"] < 2010)]
d3= earn[earn["Year"] >=2010]
```

```python
d1
```

```python
mejor_d1=d1.groupby(["Sport","Name"]).agg({"earnings ($ million)":sum}).sort_values(
        by=["earnings ($ million)"], ascending = False)
mejor_d1
#Para la primera decada el atleta con mayores ganancias es "Michael Jordan"
```

```python
mejor_d2=d2.groupby(["Sport","Name"]).agg({"earnings ($ million)":sum}).sort_values(
        by=["earnings ($ million)"], ascending = False)
mejor_d2
#Para la segunda decada el atleta con mayores ganancias es "Tiger Woods"
```

```python
mejor_d3=d3.groupby(["Sport","Name"]).agg({"earnings ($ million)":sum}).sort_values(
        by=["earnings ($ million)"], ascending = False)
mejor_d3
#Para la tercera decada el atleta con mayores ganancias es "Floyd Mayweather"
```
